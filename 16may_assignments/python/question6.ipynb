● Get all Pokemons whose spawn time is less than 300 seconds.
Note - spawn time format is "05:32”, so assume “minute: second” format and perform the analysis.

```python
import json
from urllib.request import urlopen
import pandas as pd


url = "https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json"
response = urlopen(url)

# from url in data
data_json = json.loads(response.read())
  
# print the json response
# print(data_json)

# acessing pokemon dictionary as dataframes
df = pd.DataFrame(data_json['pokemon'])
```

```python
df
```

```python
# Get all Pokemons whose spawn rate is less than 5%

#counting spawn chances in % and comparing with < 5 and rows filtered accordingly.. 

df[df["spawn_chance"]*100 < 5]
```

```python
# Get all Pokemons that have less than 4 weaknesses

#accessing weakness as list and checking length of it..
df[df["weaknesses"].apply(lambda x: len(x)<4)]
```

```python
# Get all Pokemons that have no multipliers at all

df[df["multipliers"].apply(lambda x: x is None or len(x)==0)]
```

```python
# Get all Pokemons that do not have more than 2 evolutions

## question doesnot clarify which evolution column should be consider..
# can apply same logic on both of them
import numpy as np

## all null values rows...
dd = df[df["prev_evolution"].apply(lambda x: str(x)=="nan")]

## all not null values...
dd1 = df[~ df["prev_evolution"].isin(dd["prev_evolution"])]

dd2 = dd1[dd1["prev_evolution"].apply(lambda x: len(x)<2)]

#so null/0 evaluation and 1 will be listed...
dd.append(dd2)



```

```python
# Get all Pokemons whose spawn time is less than 300 seconds.

df['spawn_time'] = df['spawn_time'].replace(np.nan, 0)
# df['sec']=df['spawn_time'].dt.total_seconds()
df["spawn_time_mm"] = df["spawn_time"].str.split(":").str[0]
df["spawn_time_ss"] = df["spawn_time"].str.split(":").str[1]

df['spawn_time_mm'] = df['spawn_time_mm'].replace('N/A', 0)
df["spawn_time_mm"] = df["spawn_time_mm"].fillna(0).astype('int')
df["spawn_time_ss"] = df["spawn_time_ss"].fillna(0).astype('int')

df["sec"] = df["spawn_time_mm"] * 60 + df["spawn_time_ss"]

df[df['sec']<300]
```

```python
#Get all Pokemon who have more than two types of capabilities

# considering capabilities as number of weakness
df[df['weaknesses'].str.len()>2]
```
