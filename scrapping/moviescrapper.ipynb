```python
def webEleToArr(WebElements):
    arr = []
    for i in WebElements:
        arr.append(i.get_attribute('href'))
        # print(i.text)
    # print(len(arr))
    return arr
```

```python
arr =webEleToArr(movie_link_list)
```

```python
titles = []
cost = []
upc = []
format =[]

for i in range(len(arr)):
    if i%2 == 0:
        movie_url = arr[i]
        # print(movie_url)
        driver.get(movie_url)
        movie_page = driver.current_window_handle
        time.sleep(10)
        
        try:
            movie_title = driver.find_element(By.XPATH, "//div[@class='product__title']/h1")
            titles.append(movie_title.text)
        except (NoSuchElementException):
             titles.append("")
        try:
            movie_cost = driver.find_element(By.XPATH, "//*[@id='price-template--16972662210778__main']/div/div/div[1]/span[2]")
            cost.append(movie_cost.text)
        except (Exception):
            cost.append("")
        
        try:    
            # print("in upc")
            found_upc = False
            found_format = False
            
            parent_div = driver.find_elements(By.XPATH, "//*[@id='ProductInfo-template--16972662210778__main']/div[7]/p")
            
            for item in parent_div:
                item.find_element(By.XPATH, "//*[@id='ProductInfo-template--16972662210778__main']/div[7]/p/strong")
                if "UPC:" in item.text:
                    found_upc = True
                    upc.append(item.text.replace("UPC:", ""))
                if "Format: " in item.text:
                    found_format = True
                    format.append(item.text.replace("Format:", ""))
                        
            if not found_upc:
                upc.append("")
            if not found_format:
                format.append("")
                         
        except (Exception):
            upc.append("")
            format.append("")
        
        
```

```python
import pandas as pd

dict = {'Movie/TV Title': titles,
        'Cost USD': cost,
        'UPC': upc,
        'Format': format}

df = pd.DataFrame(dict)
```
