```python
#most installed app...
df_cat_installs = df.groupby(['Category'])['Installs'].sum().sort_values(ascending=False).reset_index() # most installed app
```

```python
df_cat_installs
```

```python
df_cat_installs_10 = df_cat_installs.head(10)
```

```python
ax = sns.barplot(x = 'Installs' , y = 'Category' , data = df_cat_installs_10 )
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

```python
# top 5 app most installed in all category

df11 = df.groupby(['Category','App'])['Installs'].sum().reset_index() # most installed app
```

```python
df11 = df11.sort_values('Installs', ascending=False)
```

```python
df11[df11.Category == 'GAME'][:5]
```

```python
df11[df11.Category == 'PRODUCTIVITY'][:5]
```

```python
df11[df11.Category == 'FAMILY'][:5]
```

```python
 ### What are the Top 5 most installed Apps in Each popular Categories ??
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = df11[df11.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()

```
